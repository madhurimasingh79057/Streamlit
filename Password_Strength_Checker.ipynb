# Password Strength Checker

```python
%pip install streamlit
import streamlit as st
```

```python
st.title("Password Checker")

pwd=st.text_input(
    "Password",
    type="password"
)

if st.button("Check"):

    score=0

    if len(pwd)>=8:
        score+=1

    if any(c.isdigit() for c in pwd):
        score+=1

    if any(c.isupper() for c in pwd):
        score+=1

    st.write("Strength:",score,"/3")
```
