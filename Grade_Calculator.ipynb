```python
import streamlit as st
```

```python
# Heading of App
st.title("рџЋ“Smart Grade Calculator")
st.write(" Drag the sliders to change marks.No keyword needed!")
st.divider()

# Sliders(from 0 to 100) for scoring without the kayboard
# min_value =0 ,max_value=100,value=50(starting value)

maths = st.slider("вћ• Mathematics Marks", min_value=0, max_value=100, value=60)
science = st.slider("рџ”¬ Science Marks", min_value=0, max_value=100, value=65)
english = st.slider("рџ“– English Marks", min_value=0, max_value=100, value=70)
st.divider()

# Calculating Total and Percentage
total_marks = maths + science + english
percentage = total_marks / 3

# Displaying the result on the screen
st.subheader("рџ“Љ Your Result Summary")

# Display data side by side using columns
col1, col2 = st.columns(2)
col1.metric(label="Total Marks Obtained", value=f"{total_marks} / 300")
col2.metric(label="Percentage", value=f"{percentage:.2f}%")

# Deciding grades Based on Percentage
if percentage >= 90:
    grade = "A+ (Outstanding) рџЊџ"
    color = "success"
elif percentage >= 80:
    grade = "A (Excellent) рџ‘Џ"
    color = "success"
elif percentage >= 70:
    grade = "B (Very Good) рџ‘Ќ"
    color = "info"
elif percentage >= 60:
    grade = "C (Good) рџ‘Њ"
    color = "info"
elif percentage >= 50:
    grade = "D (Pass) рџ“ќ"
    color = "warning"
else:
    grade = "F (Fail) вќЊ Needs Improvement"
    color = "error"

# Displaying the Final Grade in a beautiful box
if color == "success":
    st.success(f"Final Grade: {grade}")
elif color == "info":
    st.info(f"Final Grade: {grade}")
elif color == "warning":
    st.warning(f"Final Grade: {grade}")
else:
    st.error(f"Final Grade: {grade}")
```
