```python
import pandas as pd
import streamlit as st
```

```python
# Heading of App
st.title("📊 Dynamic Table Generator")

# Getting the number of rows and columns from the user
rows = st.number_input("How many rows?", min_value=1, max_value=10, value=3)
cols = st.number_input("How many columns?", min_value=1, max_value=10, value=3)

# Creating the Generate button
if st.button("Generate Table"):

# Creating an empty Dictionary to store the data
  data={}

# creating column and row data by running a loop
  for c in range(cols):
    col_name =(f"Column{c +1}")   # The name of the column(e.g. column1)

# Enter sample data for each row within each
    data[col_name] = [f"Row {r+1},Col {c+1}" for r in range(rows)]

# Converting Dictionary to pandas DataFrame (Table)
  df = pd.DataFrame(data)

# Displaying tables beautifully on the screen
  st.subheader("Your Generated Table:")
  st.dataframe(df)
```
